```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('i.png',0)

invert=cv2.bitwise_not(img)


n=5

kernel = np.ones((n,n),np.uint8)

erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(151),plt.imshow(img),plt.title('Original')
plt.subplot(152),plt.imshow(dilation),plt.title('dilation')
plt.subplot(153),plt.imshow(erosion),plt.title('erosion')
plt.subplot(154),plt.imshow(opening),plt.title('opening')
plt.subplot(155),plt.imshow(closing),plt.title('closing')
plt.subplot(156),plt.imshow(gradient),plt.title('gradient')


plt.subplot(157),plt.imshow(kernel),plt.title('kernel')
plt.xticks([]), plt.yticks([])
```
